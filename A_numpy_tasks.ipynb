```python
# A.3 Activation Function (Sigmoid in Sales Forecasting)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
forecast_values = np.array([-2, -1, 0, 1, 2])
sigmoid_output = sigmoid(forecast_values)

print("Forecast values:")
print(forecast_values)

print("Sigmoid output:")
print(sigmoid_output)
```

```python
# A.4 Gradient Calculation (Learning Adjustment)
def sigmoid_gradient(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
gradient_output = sigmoid_gradient(forecast_values)

print("Forecast values:")
print(forecast_values)

print("Sigmoid gradient:")
print(gradient_output)
```
